```python
import pandas as pd

college = pd.read_excel('budget.xlsx', sheet_name='College')

mar_22 = college[4:38]
mar_22
```

```python
# Create F&B dataframe
```

```python
# Make first row into header

mar22 = mar_22.transpose()
mar22.columns = mar22.iloc[0]
mar22 = mar22[1:] # delete first row which is a duplicate of the header

# Keep columns that are only under F&B

fnb_mar22 = mar22.iloc[:,0:5:1] # slice by column index from position 0 to 5 with increments of 1

# Rename "Total" header to F&B and drop F&B column

fnb_mar22 = fnb_mar22.drop('F&B', axis = 1)
fnb_mar22 = fnb_mar22.rename(columns = {'Total':'F&B'})

# Move date from header to first row index and drop the other rows

fnb_mar22 = fnb_mar22.rename(index={fnb_mar22.index[0]: fnb_mar22.columns.name})
fnb_mar22.columns.name = 'Category'
fnb_mar22 = fnb_mar22.iloc[:1].reset_index()
fnb_mar22 = fnb_mar22.rename(columns={'index': 'Date'})
fnb_mar22
```

```python
# Make first row into header

may22 = may_22.transpose()
may22.columns = may22.iloc[0]
may22 = may22[1:] # delete first row which is a duplicate of the header

# Keep columns that are only under Housing

housing_may22 = may22.iloc[:,11:17:1] # slice by column index from position 0 to 5 with increments of 1

# Rename "Total" header to Housing and drop Housing column

housing_may22 = housing_may22.drop('Housing', axis = 1)
housing_may22 = housing_may22.rename(columns = {'Total':'Housing'})

# Move date from header to first row index and drop the other rows

housing_may22 = housing_may22.rename(index={housing_may22.index[0]: may_22.iloc[0]['From 03/14/22']})
housing_may22.columns.name = 'Category'
housing_may22 = housing_may22.iloc[:1].reset_index()
housing_may22 = housing_may22.rename(columns={'index': 'Date'})
housing_may22
```

```python
# Join March, April and May '22 dataframes

housing = pd.concat([housing_mar22, housing_apr22, housing_may22], ignore_index = True)
housing
```

```python
# Consolidate columns to make the Housing dataframe cleaner

housing.rename(columns={housing.columns[6]: 'Utilities_2', housing.columns[8]: 'Utilities_3'},inplace=True)
housing = housing.fillna(0)

# Other column that consist of expenses other than rent and utilities
housing['Other'] = housing['One City App Fee'] + housing['One City Deposit'] + housing['Renter Insurance (May and June)']

# Rent column
housing['Rent'] = housing['Rent'] + housing['Rent MSN'] + housing['Rent PHL']

# Utilities column
housing['Utilities'] = housing['Utilities'] + housing['Utilities_2'] + housing['Utilities_3']

housing.drop(['One City App Fee', 'One City Deposit', 'Utilities_2', 'Rent MSN', 'Utilities_3', 'Rent PHL', 'Renter Insurance (May and June)'], axis=1, inplace=True)
housing = housing[['Date', 'Rent', 'Utilities', 'Other', 'Housing']]

housing
```

```python
# Add Comment column
housing['Comment'] = ['Includes One City application fee and deposit.', None, 'Includes rent in One City.']
housing
```
